Here is some complementary analysis I did after realizing I needed while making the charts. It calculates how long the court takes to judge the political parties' financial statements

```python
#.dt.year extracts just the year from a datetime column
df_anual["Data_julgamento"] = pd.to_datetime(df_anual["Data_julgamento"], dayfirst=True)
df_anual["Ano_julgamento"] = df_anual["Data_julgamento"].dt.year
```

```python
df_anual[['Ano', 'Ano_julgamento']].head()
```

```python
#calculating the difference between the year of protocol and the year of judgement
df_anual["Diferença_anos"] = df_anual["Ano_julgamento"] - df_anual["Ano"]
```

```python
df_anual["Diferença_anos"].describe()
#takes an average of 3.8 years to judge, but it can take up to 14 years
```
